Define symbols
----

Primary
--

```python
eps = Symbol('\epsilon')
k_x, k_y, k_z = symbols('k_x, k_y, k_z')
w, c_0, c, d = symbols('\omega, c_0, c, d')
```

Secondary
--

```python
p_1, p_2, p_3, p_4 = symbols('p_1, p_2, p_3, p_4')
k_z_1, k_z_2, k_z_3, k_z_4 = symbols('k_z1, k_z2, k_z3, k_z4')
mod_k_z = Symbol('k_z')
k_1, k_2, k_3, k_4 = symbols('k_1, k_2, k_3, k_4')
c = c_0 / sqrt(eps)
```

```python
p_1, p_2, p_3, p_4, k_z_1, k_z_2, k_z_3, k_z_4, mod_k_z, k_1, k_2, k_3, k_4, c
```

Eigenmodes
---

```python
mod_k_z = sqrt(c/w - k_y**2 - k_x**2)

k_z_1 = mod_k_z
k_z_2 = -mod_k_z
k_z_3 = mod_k_z
k_z_4 = -mod_k_z

k_1 = Matrix([k_x, k_y, k_z_1])
k_2 = Matrix([k_x, k_y, k_z_2])
k_3 = Matrix([k_x, k_y, k_z_3])
k_4 = Matrix([k_x, k_y, k_z_4])

p_1 = Matrix([cos(k_x/k_z_1), 0, sin(k_x/k_z_1)])
p_2 = Matrix([cos(k_x/k_z_2), 0, sin(k_x/k_z_2)])
p_3 = Matrix([0, -cos(k_y/k_z_3), sin(k_y/k_z_3)])
p_4 = Matrix([0, cos(k_y/k_z_4), sin(k_y/k_z_4)])

# Norm
p_1 = simplify(p_1 / p_1.dot(p_1))
p_2 = simplify(p_2 / p_2.dot(p_2))
p_3 = simplify(p_3 / p_3.dot(p_3))
p_4 = simplify(p_4 / p_4.dot(p_4))

q_1 = simplify((c/w) * k_1.cross(p_1))
q_2 = simplify((c/w) * k_2.cross(p_2))
q_3 = simplify((c/w) * k_3.cross(p_3))
q_4 = simplify((c/w) * k_4.cross(p_4))
```
